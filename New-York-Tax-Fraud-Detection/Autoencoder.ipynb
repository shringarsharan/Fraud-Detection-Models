```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from scipy import stats
import tensorflow as tf
import seaborn as sns
from pylab import rcParams
from sklearn.model_selection import train_test_split
from keras.models import Model, load_model
from keras.layers import Input, Dense
from keras.callbacks import ModelCheckpoint, TensorBoard
from keras import regularizers
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
%%time
data = pd.read_csv('Data/NY data features.csv')
```

```python
data.head()
```

```python
RECORD = data['RECORD']
data.drop(columns = ['RECORD'], inplace = True)
```

```python
data.head()
```

```python
pca = PCA(n_components=6)
data_pca = pca.fit_transform(data)
```

```python
ss = StandardScaler()
data_pca = ss.fit_transform(data_pca)
```

```python
data_pca_train, data_pca_test = train_test_split(data_pca, test_size=0.2, random_state=42)
```

```python
input_dim = data_pca.shape[1]
encoding_dim = 3


input_layer = Input(shape=(input_dim, ))
encoder = Dense(encoding_dim, activation="relu", 
                activity_regularizer=regularizers.l1(10e-5))(input_layer)

decoder = Dense(encoding_dim, activation='relu')(encoder)
decoder = Dense(input_dim, activation='linear')(decoder)

autoencoder = Model(inputs=input_layer, outputs=decoder)

autoencoder.compile(optimizer='adam', 
                    loss='mean_squared_error', 
                    metrics=['mae'])
```

```python
checkpointer = ModelCheckpoint(filepath="model.h5",
                               verbose=0,
                               save_best_only=True)
tensorboard = TensorBoard(log_dir='./logs',
                          histogram_freq=0,
                          write_graph=True,
                          write_images=True)
```

```python
nb_epoch = 10
batch_size = 1024

history = autoencoder.fit(data_pca, data_pca,
                    epochs=nb_epoch,
                    batch_size=batch_size,
                    shuffle=True,
                    verbose=1).history
```

```python
predictions = autoencoder.predict(data_pca, batch_size = 1024, verbose = 1)
```

```python
predictions.shape
```

```python
mse = np.mean(np.power(data_pca - predictions, 2), axis=1) ** 0.5
```

```python
mse.shape
```

```python
data['autoencoder_score'] = mse
data['RECORD'] = RECORD 
```

```python
df_pca = pd.DataFrame(np.power(data_pca - predictions, 2)).rename(columns = {0:'PC1', 1:'PC2', 2:'PC3', 3:'PC4', 4:'PC5', 5:'PC6'})
```

```python
df_pca['autoencoder_score'] = mse
df_pca['RECORD'] = RECORD 
```

```python
df_pca.head()
```

```python
df_pca = df_pca.sort_values(by = 'autoencoder_score', ascending=False).reset_index().drop(columns = ['index']).reset_index().rename(columns = {'index': 'rank'})
```
