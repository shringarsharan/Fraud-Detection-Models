```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
from sklearn.preprocessing import StandardScaler
```

```python
%%time
df=pd.read_csv("Data/NY data no missing values.csv")
```

```python
df.head()
```

```python
df['S1']=df['LTFRONT']*df['LTDEPTH']
df['S2']=df['BLDFRONT']*df['BLDDEPTH']
df['S3']=df['S2']*df['STORIES']
```

```python
df['R1']=df['FULLVAL']/df['S1']
df['R2']=df['FULLVAL']/df['S2']
df['R3']=df['FULLVAL']/df['S3']
df['R4']=df['AVLAND']/df['S1']
df['R5']=df['AVLAND']/df['S2']
df['R6']=df['AVLAND']/df['S3']
df['R7']=df['AVTOT']/df['S1']
df['R8']=df['AVTOT']/df['S2']
df['R9']=df['AVTOT']/df['S3']   
```

```python
df['ZIP5']=df['ZIP'].apply(str).str.slice(0,5)
```

```python
df['ZIP3']=df['ZIP5'].str.slice(0,3)
```

```python
zip5_group=df.groupby(['ZIP5'])['R1','R2','R3','R4','R5','R6','R7','R8','R9'].mean().reset_index()
zip5_group=zip5_group.add_prefix('zip5_')
```

```python
zip3_group=df.groupby(['ZIP3'])['R1','R2','R3','R4','R5','R6','R7','R8','R9'].mean().reset_index()
zip3_group=zip3_group.add_prefix('zip3_')
```

```python
tax_group=df.groupby(['TAXCLASS'])['R1','R2','R3','R4','R5','R6','R7','R8','R9'].mean().reset_index()
tax_group=tax_group.add_prefix('tax_')
tax_group.head()
```

```python
b_group=df.groupby(['B'])['R1','R2','R3','R4','R5','R6','R7','R8','R9'].mean().reset_index()
b_group=b_group.add_prefix('b_')
```

```python
df['R1_avg']=df.R1.mean()
df['R2_avg']=df.R2.mean()
df['R3_avg']=df.R3.mean()
df['R4_avg']=df.R4.mean()
df['R5_avg']=df.R5.mean()
df['R6_avg']=df.R6.mean()
df['R7_avg']=df.R7.mean()
df['R8_avg']=df.R8.mean()
df['R9_avg']=df.R9.mean()
```

```python
df=pd.merge(df,zip5_group,left_on='ZIP5',right_on='zip5_ZIP5',how='left')
df=pd.merge(df,zip3_group,left_on='ZIP3',right_on='zip3_ZIP3',how='left')
df=pd.merge(df,tax_group,left_on='TAXCLASS',right_on='tax_TAXCLASS',how='left')
df=pd.merge(df,b_group,left_on='B',right_on='b_B',how='left')
```

```python
df.columns
```

```python
df['R1_zip5_scaled']=df['R1']/df['zip5_R1']
```

```python
org=['R1','R2','R3','R4','R5','R6','R7','R8','R9']
zip5=['zip5_R1', 'zip5_R2', 'zip5_R3', 'zip5_R4','zip5_R5', 'zip5_R6', 'zip5_R7', 'zip5_R8', 'zip5_R9']
d=['R1_zip5_scaled','R2_zip5_scaled','R3_zip5_scaled','R4_zip5_scaled','R5_zip5_scaled','R6_zip5_scaled','R7_zip5_scaled','R8_zip5_scaled','R9_zip5_scaled']

for i in range(len(org)):
    df[d[i]]=df[org[i]]/df[zip5[i]]
```

```python
df.columns
```

```python
to_add='zip3'

org=['R1','R2','R3','R4','R5','R6','R7','R8','R9']
zip3=[]
d=[]
for elem in org:
    elem=to_add+'_'+elem
    zip3.append(elem)

for elem in org:
    elem=elem+to_add+'_scaled'
    d.append(elem)

for i in range(len(org)):
    df[d[i]]=df[org[i]]/df[zip3[i]]
```

```python
df.columns
```

```python
to_add='b'

org=['R1','R2','R3','R4','R5','R6','R7','R8','R9']
b=[]
d=[]
for elem in org:
    elem=to_add+'_'+elem
    b.append(elem)

for elem in org:
    elem=elem+to_add+'_scaled'
    d.append(elem)

for i in range(len(org)):
    df[d[i]]=df[org[i]]/df[b[i]]
```

```python
df.columns
```

```python
to_add='tax'

org=['R1','R2','R3','R4','R5','R6','R7','R8','R9']
tax=[]
d=[]
for elem in org:
    elem=to_add+'_'+elem
    tax.append(elem)

for elem in org:
    elem=elem+to_add+'_scaled'
    d.append(elem)

for i in range(len(org)):
    df[d[i]]=df[org[i]]/df[tax[i]]
```

```python
org=['R1','R2','R3','R4','R5','R6','R7','R8','R9']
overall=[]
d=[]
for elem in org:
    elem=elem+'_avg'
    overall.append(elem)

for elem in org:
    elem=elem+'overall'+'_scaled'
    d.append(elem)

for i in range(len(org)):
    df[d[i]]=df[org[i]]/df[overall[i]]
```
